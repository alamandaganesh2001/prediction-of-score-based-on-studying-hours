# TASK-1 : Prediction using Supervised **ML**

Predicting score on the basis of hours/day spent studying.

# ***Importing libraries***

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
from sklearn.model_selection import train_test_split  
from sklearn.linear_model import LinearRegression
```

## **Importing Dataset**

```python
url="http://bit.ly/w-data"
data= pd.read_csv(url)
print(data.head())
```

# ***Plotting Dataset***

```python
data.plot(x='Hours', y='Scores', style='o')
plt.title('Hours/day and Scores')  
plt.xlabel('Hours/day spend in studying')  
plt.ylabel('Scores gained')  
plt.show()
```

```python
x=data.iloc[:,:-1].values#feature
y=data.iloc[:,1].values#label
```

# *Creating test dataset out of the total*

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=0.2, random_state=0)
```

# *Training the Model*

```python
regressor = LinearRegression()
regressor.fit(xtrain, ytrain) 
print("Your model has been trained")
```

# *Testing the Model*

Given testing value of hours=9.25, predicting the score based on the input value.


```python
hours=9.25
hours=np.array(hours).reshape(1,-1)
score_prediction=regressor.predict(hours)
print("After studying for "+str(float(hours))+" hours/day, one can score "+str('{:.2f}'.format(float(score_prediction))+"%."))
```

# Evaluating the trained *Model*

To evaluate the accuracy of the system, we need to test the system with Hours to which we aleady have **Scores**

```python
pred=regressor.predict(xtest)
```

### *Checking visually the difference between Actual and Predicted value*

```python
comparision = pd.DataFrame({'Actual Values':ytest,'Predicted Values':pred})
comparision
```

## ***Plotting the predicted values to visualize***

```python
m = regressor.coef_
c=regressor.intercept_
Y=m*xtest+c
plt.scatter(xtest,ytest)
plt.plot(xtest,Y)
plt.title('Hours/day and Scores')  
plt.xlabel('Hours/day spend in studying')  
plt.ylabel('Scores gained') 
plt.show()
```

## Calculating Mean Squared Error to make a comparision between predicted and actual values.

```python
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(ytest, pred)
result=math.sqrt(mse)
print('{:.2f}'.format(result))
```
